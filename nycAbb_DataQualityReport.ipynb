### Collect summary info for all the numeric variables

```python
col_num = df.select_dtypes(include=['float', 'int']).columns
col_num
```

```python
colname = []
datatype = []
number_records_have_value = []
percent_populated = []
number_records_have_value0 = []
value_median = []
value_mean = []
value_std = []
value_min = []
value_max = []
print('Number of numeric variables: ', len(col_num))
```

```python
for col in col_num:
    colname.append(col)
    datatype.append(df[col].dtype)
    number_records_have_value.append(df[col].count())
    percent_populated.append((round(df[col].count()*100 / len(df), 2)))
    number_records_have_value0.append(len(df[df[col] == 0]))
    value_median.append(df[col].median())
    value_mean.append(df[col].mean())
    value_std.append(df[col].std())
    value_min.append(df[col].min())
    value_max.append(df[col].max())
```

```python
my_dict = {
    'Field': colname,
    'Datatype': datatype,
    'NumRec_HaveVal': number_records_have_value,
    'PercentPopulated': percent_populated,
    'NumRec_HaveValZero': number_records_have_value0,
    'Median': value_median,
    'Mean': value_mean,
    'Stand_Dev': value_std,
    'Min': value_min,
    'Max': value_max
}
df_num_sum = pd.DataFrame(my_dict)
df_num_sum
```

#### Histogram helps to understand the distribution of values in single continuous column
A histogram gives the following information:
* What is the distribution of data? What is the max and min value in the column
* What is the central tendency of data? What is the median, mean and standard deviation of value?
* What is the skewness of data? Is the data negative skewed (outliers on the left side) or positive skewed (outliers on the right side)?

```python
h = sns.displot(df, x = col, bins = 30)

# Add a vertical line across the axes
h.ax.axvline(df[col].mean(), color='red', linestyle='dashed', linewidth=1)

# Add text
min_ylim, max_ylim = h.ax.get_ylim()
min_xlim, max_xlim = h.ax.get_xlim()
h.ax.text((max_xlim+df[col].mean())/2, max_ylim*0.9, 'Mean: {:0.2f}'.format(df[col].mean()))
```
